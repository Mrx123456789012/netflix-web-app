```python
import numpy as np 
import pandas as pd
```

```python
series=pd.read_csv("netflix.csv")
series.head(3)
```

```python
# we will now drop the columns which we don't require for our project
# we will only keep those columns which we require for the project
```

```python
# we will keep these columns only 
#title
#id
#genre
#plot
#cast
#image url - for the posters 
```

```python
netflix=series[['title','imdb_id','genres','summary','cast','image_url']]
```

```python
netflix.head(1)
```

```python
# pre-processing 
netflix.isnull().sum()
# as we can see genres column contain 2 null rows so we have to drop them
```

```python
netflix1=netflix.dropna()
df=netflix1.drop(7004)
```

```python
netflix1.isnull().sum()
```

```python
df.isnull().sum()
```

```python
# now we will check for the duplicate data
df.duplicated().sum()
```

```python
# we are now making a list of every row in our summary and genres columns
# so we can concatenate this three columns summary,genres,cast this will be
#called tags

df['summary']=df['summary'].apply(lambda x:x.split())
        
```

```python
df.head(1)
```

```python
df['genres']=df['genres'].apply(lambda x:x.split())
```

```python
df.head(1)
```

```python
df['genres']=df['genres'].apply(lambda x:[i.replace(" ","")for i in x])
```

```python
df['cast']=df['cast'].apply(lambda x:x.split())
```

```python
df['tags']=df['summary']+df['genres']+df['cast']
```

```python
new_df = df[['imdb_id','title','tags','image_url']]
```

```python
new_df
```

```python
new_df['tags']=new_df['tags'].apply(lambda x:" ".join(x))
```

```python
new_df.head()
```

```python
new_df['tags'][0]
```

```python
new_df['tags']=new_df['tags'].apply(lambda x: str(x).strip('[]').replace("'", ""))
```

```python
new_df['tags'][0]
```

```python
new_df['tags']= new_df['tags'].apply(lambda x: str(x).replace('[', '').replace(']', ''))
```

```python
new_df['tags'][0]
```

```python
new_df.head()
```

```python
new_df['tags']=new_df['tags'].apply(lambda x:x.lower())
```

```python
new_df.head()
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(text):
    y=[]
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)
```

```python
new_df.loc[:, 'tags'] = new_df['tags'].apply(stem)
```

```python
# now that we have our data in the format which we want now we will do 
# vectorization the famous technique we will use bag of words where each word in the text is considered as a feature, and the frequency of each word is counted. 
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=6000,stop_words='english')
```

```python
vectors=cv.fit_transform(new_df['tags']).toarray()
```

```python
vectors[0]
```

```python
(cv.get_feature_names())
```

```python
# now that our tags columns converted into vectors 
# we have to now calculate the distace between each movie
# we will calculate cosine distance because it gives accurate result when 
# your dimensions are very big like in our case so we will use cosine
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
diff = cosine_similarity(vectors)
```

```python
sorted(list(enumerate(diff[4])),reverse=True,key=lambda x:x[1])[1:6]
```

```python
def recommend(movie):
    movie_index = new_df[new_df['title']==movie].index[0]
    distances = diff[movie_index]
    movies_list=sorted(list(enumerate(distances)),reverse=True,key=lambda x:x[1])[1:6]
    for i in movies_list:
        print(new_df.iloc[i[0]].title)
        
```

```python
recommend('She')
```
